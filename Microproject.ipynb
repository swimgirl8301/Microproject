```python
# function that takes as input theta angle, returns time of arrival difference between a pair of detectors
#VIRGO longitude: 0.18333805213  latitude: 0.76151183984
#LIGO Handford longitude:  -2.08405676917  latitude:  0.81079526383
#LIGO Livingston longitude:  -1.58430937078 latitude:  0.53342313506
#KAGRA longitude:  2.396441015  latitude:  0.6355068497
#Longitude and latitude in radians; source LAL
#Distance between VIRGO and LIGO Handford: 8.853078 * 10^6m
#Distance between VIRGO and LIGO Livingston: 8.535732 * 10^6m
#Distance between VIRGO and KAGRA: 9.620919 * 10^6m
#Distance between LIGO Handford and LIGO Livingston: 3.027088 * 10^6m
#Distance between LIGO Handford and KAGRA: 8.047406 * 10^6m
#Distance between LIGO Livingston and KAGRA: 1.1038626 * 10^7m
#Source: https://www.omnicalculator.com/other/latitude-longitude-distance
#time = distance / velocity
velocity = 2.99792458 * 10**8

d_vlh = 8.853078 * 10**6
t_vlh = d_vlh / velocity

d_vll = 8.535732 * 10**6
t_vll = d_vll / velocity

d_vk = 9.620919 * 10**6
t_vk = d_vk / velocity

d_lhll = 3.027088 * 10**6
t_lhll = d_lhll / velocity

d_lhk = 8.047406 * 10**6
t_lhk = d_lhk / velocity

d_llk = 1.1038626 * 10**7
t_llk = d_llk / velocity

import numpy as np
import pylab as plt
import math

angles = ([])

for x in range(1000):
  theta = np.random.uniform(-math.pi, math.pi)
  angles.append(theta)

anglesTheta = np.array(angles)

times1 = ([])
for i in range(len(anglesTheta)):
  time = math.cos(anglesTheta[i])*t_vlh
  times1.append(time)



#Plot distribution for c_gw = c for VIRGO and LIGO Handford
plt.plot(anglesTheta, times1,".")
plt.title("Simulating time difference in detections between VIRGO and LIGO Handford assuming c_gw = c")
plt.xlabel("Angle Theta (rad)")
plt.ylabel("Time of arrival difference (s)")
```
